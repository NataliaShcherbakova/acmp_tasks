```python
#322
s = input()
p1 = 0
p2 = 1
k = ""
while p1 + p2 <= len(s):
    p3 = p1 + p2
    k = k + s[p3-1]
    p1 = p2
    p2 = p3
print(k)
```

```python
#691
import re
n = int(input())
for s in range(n):
    s = input()
    if len(s) == 6:
        k = re.fullmatch(r'[ABCEHKMOPTXY]\d{3}[ABCEHKMOPTXY][ABCEHKMOPTXY]', s)
        if k == None:
            print('No')
        else: print('Yes')
    else: print('No')
```
